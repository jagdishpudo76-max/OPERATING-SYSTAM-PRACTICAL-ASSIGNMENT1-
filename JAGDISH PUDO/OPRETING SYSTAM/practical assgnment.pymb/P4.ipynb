```python
def fcfs_disk(requests, head):
    seek_count = 0
    current = head

    print("FCFS Disk Scheduling")
    print("Head Movement:")

    for request in requests:
        movement = abs(request - current)
        seek_count += movement
        print(f"{current} -> {request}  (Seek = {movement})")
        current = request

    print("Total Seek Time:", seek_count)


# -------- Function Call (IMPORTANT) --------
requests = [82, 170, 43, 140, 24, 16, 190]
head = 50

fcfs_disk(requests, head)
```

```python
def fifo_page_replacement(pages, capacity):
    memory = []
    faults = 0

    print("FIFO Page Replacement")
    for page in pages:
        if page not in memory:
            faults += 1
            if len(memory) < capacity:
                memory.append(page)
            else:
                memory.pop(0)
                memory.append(page)
        print("Memory:", memory)

    print("Total Page Faults:", faults)


# -------- Function Call --------
pages = [7, 0, 1, 2, 0, 3, 0, 4]
capacity = 3

fifo_page_replacement(pages, capacity)
```

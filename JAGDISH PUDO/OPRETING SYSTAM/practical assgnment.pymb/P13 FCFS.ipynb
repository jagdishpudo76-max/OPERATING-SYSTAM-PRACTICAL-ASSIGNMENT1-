```python
def fcfs (request, head):
    total = 0
    current = head
    print("sequence:",end=" ")
    for req in request:
        print(req,end =" ")
        total += abs(current-req)
        current = req
    print("\n total head movement:",total)
request =[82,170,43,140,24,16,]
head = 50
fcfs(request,head)
```

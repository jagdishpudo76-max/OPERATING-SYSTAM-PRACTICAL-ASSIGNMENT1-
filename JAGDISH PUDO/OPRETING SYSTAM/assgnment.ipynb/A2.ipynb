```python
def fcfs(processes, burst_time):
    waiting_time = [0] * len(processes)
    for i in range(1, len(processes)):
        waiting_time[i] = waiting_time[i-1] + burst_time[i-1]

    print("FCFS Scheduling")
    for i in range(len(processes)):
        print(f"Process {processes[i]} Waiting Time: {waiting_time[i]}")


def sjf(processes, burst_time):
    data = list(zip(processes, burst_time))
    data.sort(key=lambda x: x[1])

    print("\nSJF Scheduling")
    waiting_time = 0
    for p, bt in data:
        print(f"Process {p} Waiting Time: {waiting_time}")
        waiting_time += bt


processes = [1, 2, 3]
burst_time = [10, 5, 8]

fcfs(processes, burst_time)
sjf(processes, burst_time)
```

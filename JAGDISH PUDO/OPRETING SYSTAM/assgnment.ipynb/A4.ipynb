```python
# FCFS Disk Scheduling Algorithm
def fcfs_disk(requests, head):
    seek_time = 0
    current = head

    print("FCFS Disk Scheduling Order:")
    for req in requests:
        print(current, "->", req)
        seek_time += abs(req - current)
        current = req

    print("Total Seek Time (FCFS):", seek_time)


# SSTF Disk Scheduling Algorithm
def sstf_disk(requests, head):
    seek_time = 0
    current = head
    pending = requests.copy()

    print("\nSSTF Disk Scheduling Order:")
    while pending:
        nearest = min(pending, key=lambda x: abs(x - current))
        print(current, "->", nearest)
        seek_time += abs(nearest - current)
        current = nearest
        pending.remove(nearest)

    print("Total Seek Time (SSTF):", seek_time)


# Main Program
requests = [98, 183, 37, 122, 14, 124, 65, 67]
head = 53

print("Initial Head Position:", head)
print("Disk Requests:", requests)

fcfs_disk(requests, head)
sstf_disk(requests, head)
```
